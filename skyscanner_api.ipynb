```python
import pandas as pd
```

```python
country = 'US'
currency = 'USD'
destinationplace = 'anywhere'
inboundpartialdate = 'anytime' # The return date. Format “yyyy-mm-dd”, “yyyy-mm” or “anytime”. Use empty string for oneway trip.
outboundpartialdate = 'anytime' # The outbound date. Format “yyyy-mm-dd”, “yyyy-mm” or “anytime”.
locale = 'en-US'
originplace = 'CMH'
```

```python
import requests

url = f"https://skyscanner-skyscanner-flight-search-v1.p.rapidapi.com/apiservices/browseroutes/v1.0/{country}/{currency}/{locale}/{originplace}/{destinationplace}/{outboundpartialdate}/{inboundpartialdate}"

headers = {
    'x-rapidapi-host': "skyscanner-skyscanner-flight-search-v1.p.rapidapi.com",
    'x-rapidapi-key': "2d32d04dd5msh5cb7c5e39103bebp1cd117jsn0f0dea8441dc"
    }

response = requests.request("GET", url, headers=headers).json()

response
```

```python
places_json = response["Places"]

places_df = pd.DataFrame(places_json)
places_df
```

```python
carriers_json = response["Carriers"]

carrier_map = {}
for item in carriers_json:
  carrier_map[item['CarrierId']] = item['Name']

carrier_map
```

```python
quotes_json = response["Quotes"]
quotes_json

for quote in quotes_json:
  print(quote['InboundLeg'])
```

```python
nice_quotes = []


for quote in quotes_json:

  # -- Parse outbound leg
  # Carrier
  carriers = []
  for carrier in quote['OutboundLeg']['CarrierIds']:
    carriers.append(carrier_map[carrier])

  # Origin
  outbound_origin_name = places_df[places_df['PlaceId'] == quote['OutboundLeg']['OriginId']][['Name']].to_dict('r')[0]['Name']
  outbound_origin_code = places_df[places_df['PlaceId'] == quote['OutboundLeg']['OriginId']][['IataCode']].to_dict('r')[0]['IataCode']
  
  # Destination
  outbound_destination_name = places_df[places_df['PlaceId'] == quote['OutboundLeg']['DestinationId']][['Name']].to_dict('r')[0]['Name']
  outbound_destination_code = places_df[places_df['PlaceId'] == quote['OutboundLeg']['DestinationId']][['IataCode']].to_dict('r')[0]['IataCode']

  # Departure Date 'DepartureDate'
  outbound_departure_date = quote['OutboundLeg']['DepartureDate']

  outbound_details =  {'Carriers': carriers, 'OriginName': outbound_origin_name, 'OriginCode': outbound_origin_code, 'DestinationName': outbound_destination_name, 'DestinationCode': outbound_destination_code, 'DepartureDate': outbound_departure_date}

  # -- Parse inbound leg
  # Carrier
  carriers = []
  for carrier in quote['InboundLeg']['CarrierIds']:
    carriers.append(carrier_map[carrier])

  # Origin
  inbound_origin_name = places_df[places_df['PlaceId'] == quote['InboundLeg']['OriginId']][['Name']].to_dict('r')[0]['Name']
  inbound_origin_code = places_df[places_df['PlaceId'] == quote['InboundLeg']['OriginId']][['IataCode']].to_dict('r')[0]['IataCode']
  
  # Destination
  inbound_destination_name = places_df[places_df['PlaceId'] == quote['InboundLeg']['DestinationId']][['Name']].to_dict('r')[0]['Name']
  inbound_destination_code = places_df[places_df['PlaceId'] == quote['InboundLeg']['DestinationId']][['IataCode']].to_dict('r')[0]['IataCode']

  # Departure Date 'DepartureDate'
  inbound_departure_date = quote['InboundLeg']['DepartureDate']

  inbound_details =  {'Carriers': carriers, 'OriginName': inbound_origin_name, 'OriginCode': inbound_origin_code, 'DestinationName': inbound_destination_name, 'DestinationCode': inbound_destination_code, 'DepartureDate': inbound_departure_date}

  # print quote, price, direct, outbound leg details, inbound leg details, and quote time
  nice_quotes.append([quote['QuoteId'], quote['MinPrice'], quote['Direct'], outbound_details, inbound_details, quote['QuoteDateTime']])



quote_df = pd.DataFrame(nice_quotes, columns=['QuoteId', 'MinPrice', 'Direct', 'OutboundLeg', 'InboundLeg', 'QuoteDateTime'])
quote_df
```
